```
import requests
url = "https://www.values.com/inspirational-quotes/"
response = requests.get(url)
print(response)
print(type(response))
print("---")
print(response.content)
print("---")
print(type(response.content))
```

```
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.content, 'html5lib') # seems like the soup is separated by < > so we can refer to items
# in example: <h1>  <ul> <p> text </p> </ul> </h1> as txt = item.h1.ul.text
# or example: <h1> <a src="www.me.com" class="my class"> my site </a> </h1> as site= item.h1.a['src']
print(soup) # type bs4.BeautifulSoup 
```

```

print(soup.prettify()) #type string 
```

```
table = soup.find('div', attrs={'id':'all_quotes'}) # here all_quotes refers to all quotes info, notice find not findAll
print(table) # type bs4.element.Tag 
```

```
print(table.prettify()) #type string
```

```
table2 = table.findAll('div', attrs={'class':'col-6 col-lg-4 text-center margin-30px-bottom sm-margin-30px-top'}) # notice findall
print(table2) # type bs4.element.ResultSet 
```

```
#can't run below because of find_all use
#print(table2.prettify()) 
```

```
table2[0] # same as row below ...of type bs4.element.Tag ...can prettify
print(table2[0])
print(len(table2[0])) # notice same as size of row and size of list1 below
```

```
list1 = [] # definining an array
for row in table2: # for items separated by < in list table2 ...list of bs4.element.Tag...bs4.element.ResultSet
  quote_info={} # defining a dictionary
  quote_info['theme'] = row.h5.text # row with attr h5 and content type text
  quote_info['url'] = row.a['href'] # row with attr a and option href
  quote_info['img'] = row.img['src'] # row with attr img and option src
  quote_info['quote'] = row.img['alt'].split(' #')[0] # row with attr img and option alt split by # first part
  quote_info['author'] = row.img['alt'].split(' #')[1] # row with attr img and option alt split by # 2nd part
  list1.append(quote_info)
```

```
print(list1[0]) # notice though list1[0] and row/table2[31] are same size...they are not the same type
print(row) # the last 'row' after the loop exits
```
